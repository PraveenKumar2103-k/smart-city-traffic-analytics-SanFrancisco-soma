```python
import warnings
warnings.filterwarnings("ignore", message="Could not infer format")
```

```python
import pandas as pd
```

```python
accidents = pd.read_csv("accidents_soma.csv")
```

```python
complaints = pd.read_csv("complaints_soma.csv")
```

```python
accidents.head()
```

```python

#  structure of accidents_soma
print("Accidents Data Overview:")
accidents.info()

#  structure of complaints-soma
print("\nComplaints Data Overview:")
complaints.info()
```

```python
# Preview accidents data
accidents.head()

# Preview complaints data
complaints.head()
```

```python
# Check null values in accidents data
print("Null values in accidents:")
print(accidents.isnull().sum())

# Check null values in complaints data
print("\nNull values in complaints:")
print(complaints.isnull().sum())
```

```python
# Optionally drop the 'Filed Online' column if you don't need it
accidents = accidents.drop(columns=["Filed Online"])

# Just to be sure: Drop rows missing lat/lon (none in your case, but good to include)
accidents = accidents.dropna(subset=["Latitude", "Longitude"])

# Confirm
print("Accidents cleaned nulls ✅")
```

```python
# Drop irrelevant columns
complaints = complaints.drop(columns=["Media URL", "BOS_2012"])

# Drop rows missing critical location fields
complaints = complaints.dropna(subset=["Latitude", "Longitude"])

# Confirm
print("Complaints cleaned nulls ✅")
```

```python
# Remove duplicates
accidents = accidents.drop_duplicates()
complaints = complaints.drop_duplicates()

print("Duplicates removed from both datasets ✅")
```

```python
print("Nulls after cleaning - Accidents")
print(accidents.isnull().sum())

print("\nNulls after cleaning - Complaints")
print(complaints.isnull().sum())
```

```python
accidents['Incident Datetime'] = pd.to_datetime(accidents['Incident Datetime'])
accidents['Incident Date'] = pd.to_datetime(accidents['Incident Date']).dt.date

# This line handles mixed or missing formats gracefully
accidents['Incident Time'] = pd.to_datetime(accidents['Incident Time'], errors='coerce').dt.time

accidents['Report Datetime'] = pd.to_datetime(accidents['Report Datetime'])
```

```python
print(accidents[['Incident Datetime', 'Incident Date', 'Incident Time', 'Report Datetime']].head())

print(complaints[['Opened', 'Closed', 'Updated']].head())
```

```python
# Convert to datetime first
complaints['Opened'] = pd.to_datetime(complaints['Opened'], errors='coerce')

# Now extract parts
complaints['Year'] = complaints['Opened'].dt.year
complaints['Month'] = complaints['Opened'].dt.month
complaints['DayOfWeek'] = complaints['Opened'].dt.day_name()
```

```python
print("Accidents data types:")
print(accidents[['Latitude', 'Longitude']].dtypes)

print("\nComplaints data types:")
print(complaints[['Latitude', 'Longitude']].dtypes)
```

```python
accidents['Latitude'] = pd.to_numeric(accidents['Latitude'], errors='coerce')
accidents['Longitude'] = pd.to_numeric(accidents['Longitude'], errors='coerce')

complaints['Latitude'] = pd.to_numeric(complaints['Latitude'], errors='coerce')
complaints['Longitude'] = pd.to_numeric(complaints['Longitude'], errors='coerce')
```

```python
accidents = accidents.dropna(subset=['Latitude', 'Longitude'])
complaints = complaints.dropna(subset=['Latitude', 'Longitude'])
```

```python
accidents['Location'] = accidents['Latitude'].astype(str) + "," + accidents['Longitude'].astype(str)
complaints['Location'] = complaints['Latitude'].astype(str) + "," + complaints['Longitude'].astype(str)
```

```python
print("Sample accidents location:")
print(accidents[['Latitude', 'Longitude', 'Location']].head())

print("\nSample complaints location:")
print(complaints[['Latitude', 'Longitude', 'Location']].head())
```

```python
# Round lat/lon to 3 decimal places (approx ~100m radius)
accidents['lat_round'] = accidents['Latitude'].round(3)
accidents['lon_round'] = accidents['Longitude'].round(3)

complaints['lat_round'] = complaints['Latitude'].round(3)
complaints['lon_round'] = complaints['Longitude'].round(3)
```

```python
# Perform inner merge (only locations that exist in both)
merged_data = pd.merge(
    accidents,
    complaints,
    on=['lat_round', 'lon_round'],
    suffixes=('_accident', '_complaint')
)
```

```python
print("Merged Data Sample:")
print(merged_data[['Incident Datetime', 'Opened', 'Incident Category', 'Request Type', 'Location_accident', 'Location_complaint']].head())

print("\nShape of merged data:", merged_data.shape)
```

```python
merged_data.to_csv("merged_accidents_complaints.csv", index=False)
```
